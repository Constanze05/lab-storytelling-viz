### Factor 1: The gross living area

#### The sales price for a house increases if the gross living area increases.
There are just some houses that have been sold with a big area. 
Above average houses in terms of size are also fewer.
The average area seems to be around 2000 or less and the price range differs from 1 million up to 5 million.

```python
plt.scatter(df['GrLivArea'], df['SalePrice'], color='orange')
plt.title('gross living area vs sale price', fontsize=14)
plt.xlabel('gross living area', fontsize=14)
plt.ylabel('sale price', fontsize=14)
plt.show()
```

### Factor 2: The garage area

#### The sales price for a house increases with a bigger garage.

```python
plt.scatter(df['GarageArea'], df['SalePrice'], color='red')
plt.title('garage area vs sale price', fontsize=14)
plt.xlabel('garage area', fontsize=14)
plt.ylabel('sale price', fontsize=14)
plt.show()
```
